📊 Topic 3 — Bar Chart
Bar Chart = Different different things compare . Har cheez ki ek bar (rectangle) hoti hai — jitni badi bar, utni badi value.

Real life examples - Ipl teams total runs
Phone brands sales
students marks compare
Countries population


Type 1 → Vertical Bar Chart     (sabse common)

Type 2 → Horizontal Bar Chart   (jab labels lambe hon)

Type 3 → Grouped Bar Chart      (2-3 groups compare)

Type 4 → Stacked Bar Chart      (parts of whole)

Type 5 → Bar + Value Labels     (bars pe numbers dikhao)

```python
# Type 1 - VERTICAL BAR CHART
import matplotlib.pyplot as plt
companies = ['Amazon','Flipkart','Meesho','Myntra','Snapdeal']
revenue=[450,320,180,240,90]

fig,ax =plt.subplots(figsize=(9,5))
ax.bar(companies,revenue)
#   ↑
# ax.bar bar chart create
# first argument = categories (x axis )
# second argument = values (y axis )
ax.set_title("E-commerce Revenue Comparison",fontweight='bold')
ax.set_xlabel("Companies",fontsize=12,color='purple')
ax.set_ylabel("Revenue ( Crores )",fontsize=12)


plt.tight_layout()
plt.show()
```

```python
# Step 2 -  Proper styling , colors , width , edge
import matplotlib.pyplot as plt
companies = ['Amazon', 'Flipkart', 'Meesho', 'Myntra', 'Snapdeal']
revenue   = [450, 320, 180, 240, 90]
fig, ax = plt.subplots(figsize=(9, 5))
ax.bar(companies, revenue,
       color=['#FF9900', '#F7464A', '#9B59B6', '#E91E8C', '#3498DB'],
       width=0.5,
       edgecolor='black',
       linewidth=0.8,
)
ax.set_title("E commerce comparison",fontsize=15,fontweight='bold')
ax.set_xlabel("Company",fontsize=12)
ax.set_ylabel("Revenue",fontsize=12)
ax.grid(True,axis='y',linestyle='--',alpha=0.5)


# axis = y - only horizontal grid line

plt.tight_layout()
plt.show()
```

```python
# Step - 03 VALUE LABELS ( NUMBERS IN BARS ) imp
import matplotlib.pyplot as plt
companies = ['Amazon', 'Flipkart', 'Meesho', 'Myntra', 'Snapdeal']
revenue   = [450, 320, 180, 240, 90]
fig,ax =plt.subplots(figsize=(10,7))
bars = ax.bar(companies, revenue,
              color=['#FF9900', '#F7464A', '#9B59B6', '#E91E8C', '#3498DB'],
              width=0.5,
              edgecolor='black',
              linewidth=0.8)
for bar in bars:
  height = bar.get_height()
  #  get_height = find that bar height value
  ax.text(bar.get_x() + bar.get_width() / 2,
          height + 5,
          str(height),
          ha='center',
          va='bottom',
          fontsize=11,
          fontweight='bold')

ax.set_title("E-Commerce Revenue with Values", fontsize=15, fontweight='bold')
ax.set_xlabel("Company", fontsize=12)
ax.set_ylabel("Revenue (Crores)", fontsize=12)
ax.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Step 4- HORIZONTAL CHART
import matplotlib.pyplot as plt
languages = ['Python', 'JavaScript', 'Java', 'C++', 'Kotlin', 'Swift', 'Rust']
popularity = [68, 63, 51, 44, 32, 28, 22]
fig, ax = plt.subplots(figsize=(9, 6))

bars = ax.barh(languages,popularity,
               color='steelblue',
               edgecolor='black',
               linewidth=0.6,
               height=0.7)

# value labels in horizontal
for bar in bars:
  width = bar.get_width()
  ax.text(width+0.5, bar.get_y()+bar.get_height()/2,f'{width}%',
          va='center',
            fontsize=10,
            fontweight='bold'

          )
  ax.set_title("Most Popular Programming Languages 2024",
             fontsize=14, fontweight='bold')
ax.set_xlabel("Popularity (%)", fontsize=12)
ax.set_ylabel("Language", fontsize=12)
ax.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Grouped Bar Chart
import matplotlib.pyplot as plt
import numpy as np
students = ['Rahul', 'Priya', 'Amit', 'Sneha']
math     = [78, 85, 72, 91]
science  = [82, 79, 88, 75]
english  = [70, 92, 65, 88]
x= np.arange(len(students))

width = 0.25
fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(x - width,       math,    width, label='Math',    color='royalblue',  edgecolor='black', linewidth=0.6)
bar2 = ax.bar(x,               science, width, label='Science', color='tomato',     edgecolor='black', linewidth=0.6)
bar3 = ax.bar(x + width,       english, width, label='English', color='seagreen',   edgecolor='black', linewidth=0.6)


for bar in [bar1, bar2, bar3]:
    for b in bar:
        ax.text(b.get_x() + b.get_width()/2,
                b.get_height() + 0.5,
                str(b.get_height()),
                ha='center', va='bottom',
                fontsize=8, fontweight='bold')

ax.set_xticks(x)
ax.set_xticklabels(students,fontsize=11)

ax.set_title("Student Marks Comparison", fontsize=15, fontweight='bold')
ax.set_xlabel("Students", fontsize=12)
ax.set_ylabel("Marks", fontsize=12)
```

```python
# Stacked BAR CHARt
import matplotlib.pyplot as plt
import numpy as np

quarters  = ['Q1', 'Q2', 'Q3', 'Q4'] # x axis names
product_A = [120, 150, 180, 200]         # A  values
product_B = [80,  100, 90,  130]         # B values
product_C = [50,  60,  75,  95]           # c values


x =np.arange(len(quarters))
# len(quarteres=4)
# [0,1,2,3,4] positions


width=0.5
fig, ax = plt.subplots(figsize=(9, 6)) # canvas + graph

# layer 1 lower
bar1=ax.bar(x,
            product_A,
            width,label='ProductA',color='teal',edgecolor='black',linewidth=0.6

            )

# layer 2 ( center )
bar2=ax.bar(x,product_B,width,bottom=product_A,label='Product B',
              color='tomato',
              edgecolor='black',
              linewidth=0.6)

bar3 = ax.bar(x,
              product_C,   # height [50,60,75,95]
              width,
              bottom=[a+b for a, b in zip(product_A, product_B)],
              # zip  pairs: (120,80),(150,100),(180,90),(200,130)
              # a+b = [200, 250, 270, 330]
              # Q1 - 200 start, in Q2 250
              label='Product C',
              color='seagreen',
              edgecolor='black',
              linewidth=0.6)
ax.set_xticks(x)
ax.set_xticklabels(quarters,fontsize=12) # at labels in x axis q1 q2


ax.set_title("Quarterly Sales — Stacked", fontsize=15, fontweight='bold')
ax.set_xlabel("Quarter", fontsize=12)
ax.set_ylabel("Sales (units)", fontsize=12)
ax.legend(fontsize=11) # all three


#ax.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
