TOPIC - 04 HISTOGRAM

histogram - data distribution show ex- 50 students marks
how many students between 0-10,10-20 count work of histo

BIN - a range or bucket
example 1- bin 1- between  50 to 60 studentts
bin 2 - 60 - 70 between 60-70 students

```python
import matplotlib.pyplot as plt
marks = [45, 67, 82, 55, 91, 73, 68, 55, 79, 88,
         62, 71, 55, 84, 93, 77, 65, 58, 72, 86]
fig,ax = plt.subplots(figsize=(8,5))
ax.hist(marks,bins=10)

# marks - data , bins = 5
plt.show()
```

```python
# Step - 2 Bins effect
import matplotlib.pyplot as plt

marks = [45, 67, 82, 55, 91, 73, 68, 55, 79, 88,
         62, 71, 55, 84, 93, 77, 65, 58, 72, 86]

# compare two graphs
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))


# left graph
ax1.hist(marks,bins=3,color='royalblue',edgecolor='black')
ax1.set_title("bins=3 (Broad Picture)")
ax1.set_xlabel("Marks")
ax1.set_ylabel("Students Count")


# right graph more bins
ax2.hist(marks, bins=10, color='tomato', edgecolor='black')
ax2.set_title("bins=10 (Detailed Picture)")
ax2.set_xlabel("Marks")
ax2.set_ylabel("Students Count")


plt.tight_layout()
plt.show()
```

left side - bins-3 only 3 bars broad rabges not too much detail

right graph (bins=10)
10 bars - narrow range
more bars tell us  how many students will range

# commonly 10-20 bins

```python
# Step 3 - PROPER STYLED HISTOGRAM
import matplotlib.pyplot as plt

marks = [45, 67, 82, 55, 91, 73, 68, 55, 79, 88,
         62, 71, 55, 84, 93, 77, 65, 58, 72, 86]

fig, ax = plt.subplots(figsize=(9, 5))

ax.hist(marks,bins=7,color='steelblue',edgecolor='black',linewidth=0.8,alpha=0.85)

ax.set_title("Student Marks Distribution ",fontsize=15,fontweight='bold')

ax.set_xlabel("Marks", fontsize=12)
ax.set_ylabel("Number of Students", fontsize=12)
ax.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# STEP 4 - AVERAGE LINE (axvline)
import matplotlib.pyplot as plt

marks = [45, 67, 82, 55, 91, 73, 68, 55, 79, 88,
         62, 71, 55, 84, 93, 77, 65, 58, 72, 86]

fig, ax = plt.subplots(figsize=(9, 5))

ax.hist(marks, bins=7, color='steelblue',
        edgecolor='black', linewidth=0.8, alpha=0.85)

# average find
average=sum(marks)/len(marks)
print(f"Average: {average}")

# axvline = vertical line ( line chart axhline is horizontal)
ax.axvline(x=average,
           color='red',
           linestyle='--',
           linewidth=2,
           label=f'Average: {average:.1f}')


ax.set_title("Marks Distribution with Average", fontsize=15, fontweight='bold')
ax.set_xlabel("Marks", fontsize=12)
ax.set_ylabel("Number of Students", fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# STEP 5- DATA RANDOM WITH NUMPY
import matplotlib.pyplot as plt
import numpy as np

# numpy se random data banao — 200 students ke marks
np.random.seed(42)
# seed = random numbers fix karta hai
# matlab har baar same numbers aayenge

marks = np.random.randint(30, 100, 200)
# randint(low, high, size)
# 30 se 100 ke beech, 200 random numbers

fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(marks,
        bins=15,
        color='mediumpurple',
        edgecolor='black',
        linewidth=0.7,
        alpha=0.85,
        label='Students'
        )

average = np.mean(marks)
# np.mean() = average nikalta hai — sum/count automatically

ax.axvline(x=average, color='red', linestyle='--',
           linewidth=2.5, label=f'Average: {average:.1f}')

ax.set_title("200 Students Marks Distribution", fontsize=15, fontweight='bold')
ax.set_xlabel("Marks", fontsize=12)
ax.set_ylabel("Number of Students", fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
