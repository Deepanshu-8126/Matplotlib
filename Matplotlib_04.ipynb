```python
# step 1 -
import matplotlib.pyplot as plt

# study hours marks

study_hours = [2, 4, 6, 8, 10, 3, 7, 5, 9, 1]
marks       = [45, 60, 75, 85, 95, 50, 80, 65, 90, 35]

fig,ax= plt.subplots(figsize=(8,5))

# scatter plot
ax.scatter(study_hours,marks)

# first argument - x axis
# second argument - y axis
ax.set_title("Study Hours vs Marks")
ax.set_xlabel("Study Hours")
ax.set_ylabel("Marks")
plt.tight_layout()
plt.show()

# dots are student
# study hours increase marks increase
```

```python
# step 2- styling
import matplotlib.pyplot as plt

study_hours = [2, 4, 6, 8, 10, 3, 7, 5, 9, 1]
marks       = [45, 60, 75, 85, 95, 50, 80, 65, 90, 35]

fig, ax = plt.subplots(figsize=(8, 5))


ax.scatter(study_hours,marks,color='royalblue',s=100,alpha=0.7,edgecolor='black',linewidth=0.8)

# s - size of dots
# default = 20 small , aplha- transparency when dots overlap so both show
# linewidth = border width


ax.set_title("Study Hours vs Marks", fontsize=15, fontweight='bold')
ax.set_xlabel("Study Hours", fontsize=12)
ax.set_ylabel("Marks", fontsize=12)
ax.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
# step 3- diffrent colors
import matplotlib.pyplot as plt

# 2 groups - boys and girls
boys_hours  = [2, 4, 6, 8, 10, 3]
boys_marks  = [45, 60, 75, 85, 95, 50]

girls_hours = [5, 7, 9, 1, 6, 8]
girls_marks = [70, 82, 91, 38, 78, 88]

fig, ax = plt.subplots(figsize=(9, 6))

# boys scatter

ax.scatter(boys_hours, boys_marks,
           color='royalblue',
           s=100,
           alpha=0.8,
           edgecolor='black',
           linewidth=0.7,
           label='Boys'
           )


# girls scatter
ax.scatter(girls_hours, girls_marks,
           color='tomato',
           s=100,
           alpha=0.8,
           edgecolor='black',
           linewidth=0.7,
           label='Girls'
           )
ax.set_title("Study Hours vs Marks — Boys vs Girls",
             fontsize=15, fontweight='bold')
ax.set_xlabel("Study Hours", fontsize=12)
ax.set_ylabel("Marks", fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

cities      = ['Delhi', 'Mumbai', 'Bangalore']
temperature = [38, 32, 27]
ice_cream   = [500, 380, 250]
population  = [300, 250, 150]

fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(temperature, ice_cream,
           s=population,
           color='orange',
           edgecolor='black',
           alpha=0.7)

plt.show()

# in this code we dont kow which dot denoting which city
# dot is (38,500) - delhi so wrtie delhi so we use annotate

# ax.annotate("Delhi" --- what we have to write
#              ,38,500) where write (x,y)


# at every dot write name
for i , city in enumerate(cities):
  # i=0 , city= 'delhi'
   ax.annotate(city,(temperature[i],ice_cream[i]),
              # for delhi -- (38,500)
              # for mummbai (32,380)

              xytext=(8,5),
              # aside from dot 9 - little bit right
              # 5- upper
              textcoords='offset points',
              # take xytext as iin the form of distance
fontsize=10,fontweight='bold'
  )
```

```python
# step 4- control the size of dot using data

# 6 cities
cities      = ['Delhi', 'Mumbai', 'Bangalore', 'Chennai', 'Kolkata', 'Pune']
temperature = [38, 32, 27, 35, 30, 29]
# x axis tempreature

ice_cream_sales = [500, 380, 250, 450, 300, 270]
# y axis

population = [300, 250, 150, 120, 180, 100]

# population according size of dot
# big city  = bog dot


fig, ax = plt.subplots(figsize=(11, 7))

scatter = ax.scatter(temperature, ice_cream_sales,s=population,
                     color='orange',
                     alpha=0.7,
                     edgecolor='black',linewidth=1)


# at every dot city name
for i , city in enumerate(cities):
  # enumerate - index also give
  ax.annotate(city,
              (temperature[i],ice_cream_sales[i]),
              textcoords='offset points',
              xytext=(9,6),
              fontsize=9)

ax.set_title("Temperature vs Ice Cream Sales", fontsize=15, fontweight='bold')
ax.set_xlabel("Temperature (°C)", fontsize=12)
ax.set_ylabel("Ice Cream Sales", fontsize=12)
ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# step 5- trend line
import matplotlib.pyplot as plt
import numpy as np

study_hours = [2, 4, 6, 8, 10, 3, 7, 5, 9, 1]
marks       = [45, 60, 75, 85, 95, 50, 80, 65, 90, 35]

fig, ax = plt.subplots(figsize=(9, 6))

# Scatter plot
ax.scatter(study_hours, marks,
           color='royalblue',
           s=100,
           alpha=0.8,
           edgecolor='black',
           linewidth=0.7,
           label='Students',
           zorder=5
           # ↑ zorder = who come up
           # big number = comes top
           # dots line upside
           )

#  Trend Line (Best Fit Line)
m, c = np.polyfit(study_hours, marks, 1)

# np.polyfit = best fit line finds
# first argument = X data
# second argument = Y data
# third argument = 1 (straight line)
# return  → m (slope) annd c (intercept)
# line equation: y = mx + c



x_line = np.linspace(min(study_hours), max(study_hours), 100)
# np.linspace = min se max tak 100 equally spaced points
# min(study_hours) = 1
# max(study_hours) = 10
# x_line = [1.0, 1.09, 1.18, ... 10.0] (100 points)

y_line = m * x_line + c
# har x point ke liye y calculate karo
# y = mx + c formula

ax.plot(x_line, y_line,
        color='red',
        linewidth=2,
        linestyle='--',
        label='Trend Line',
        zorder=3
        )

ax.set_title("Study Hours vs Marks with Trend Line",
             fontsize=15, fontweight='bold')
ax.set_xlabel("Study Hours", fontsize=12)
ax.set_ylabel("Marks", fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
