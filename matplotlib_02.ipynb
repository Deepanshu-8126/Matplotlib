```python
# LINE PLOT -- According to time changes
# example - January to december tempreature - line plot
# comparing - bar chart
```

```python
# SINGLE LINE PLOT
import matplotlib.pyplot as plt
# 12 months data
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']


temp = [15, 17, 22, 28, 35, 40, 38, 37, 32, 26, 19, 14]

# wide graph less hieght
fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(months, temp,
        color='tomato',          # tomato =  shade of red/orange
        linewidth=2.5,           # line 2.5 unit
        marker='o',              # circle marker
        markersize=8,            # circle size 8
        markerfacecolor='white', # circle inside white color
        markeredgecolor='tomato',# circle  BORDER tomato color
        markeredgewidth=2,       # border
        label='Temperature'      # legend name
        )

ax.set_title("Delhi Monthly Temperature", fontsize=15, fontweight='bold')
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Temperature (°C)", fontsize=12)

ax.legend()
# legend-- a box in which what the line meaning
ax.grid(True,linestyle='--',alpha=0.5)
plt.xticks(rotation=45)

# x ticks x axis labels
# rotation 45 degree - overlap not happened

plt.tight_layout()
plt.show()

```

```python
# STEP - 03 fil_betweeen(AREA FILL)
import matplotlib.pyplot as plt
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']
temp = [15, 17, 22, 28, 35, 40, 38, 37, 32, 26, 19, 14]

fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(months, temp, color='tomato', linewidth=2.5,
        marker='o', markerfacecolor='white',
        markeredgecolor='tomato', markeredgewidth=2)


# fill between   (color fill) - color fill at lower
ax.fill_between(months,
                temp,
                alpha=0.2, # opacity
                color='tomato')

# opacity -- alpha
ax.set_title("Delhi Tempreature - Area Fill",fontsize=15,fontweight='bold')
ax.set_xlabel('Month',fontsize=12)
ax.set_ylabel('Tempreature(°c)',fontsize=12)
ax.grid(True,linestyle='--',alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

MULTIPLE LINES (2-3 LINES IN ONE GRAPH)

```python
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']
delhi   = [15, 17, 22, 28, 35, 40, 38, 37, 32, 26, 19, 14]
mumbai  = [24, 25, 27, 30, 33, 32, 30, 29, 30, 31, 28, 25]
shimla  = [5,  6,  10, 15, 19, 22, 21, 20, 17, 12, 7,  4 ]

fig, ax = plt.subplots(figsize=(12, 6))

# all three lines plot differently
#  but one ax.plot() for this three

# 01 - months , delhi

ax.plot(months,delhi,color='tomato',marker='o',linewidth=2,label='Delhi')

# 02 - months , mumbai

ax.plot(months, mumbai,
        color='royalblue', marker='s', linewidth=2, label='Mumbai')

ax.plot(months,shimla,
        color='seagreen',marker='^',linewidth=2,label='Shimla')


ax.set_title("Temperature Comparison — 3 Cities", fontsize=15, fontweight='bold')
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Temperature (°C)", fontsize=12)
ax.legend(fontsize=11)
# All three name in legned - delhi , mumbai , shimla will be shown


ax.grid(True, linestyle='--', alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Step - 05 Average line show

months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']
temp = [15, 17, 22, 28, 35, 40, 38, 37, 32, 26, 19, 14]

fig, ax = plt.subplots(figsize=(11, 5))

ax.plot(months, temp, color='tomato', marker='o',
        linewidth=2.5, markerfacecolor='white',
        markeredgecolor='tomato', label='Temperature')


# Average find
average = sum(temp)/len(temp)
# sum = 323 len(temp)=12 average = 323/12=26.9


#axhline = horizontal line in whole width
ax.axhline(y=average,
           color='navy',
           linestyle='--',
           linewidth=1.8,
           label=f"Average:{average:.1f}°C")


ax.set_title("Delhi Temperature with Average Line", fontsize=15, fontweight='bold')
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Temperature (°C)", fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, linestyle='--', alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



# axhline use - average , target , threshold
```

```python
# practice
import matplotlib.pyplot as plt
Subjects =['Math', 'English', 'OS', 'Python', 'DBMS']
Marks = [60,70,67,78,78]
fig, ax = plt.subplots(figsize=(11, 5))

ax.plot(Subjects, Marks, color='green', marker='o',
        linewidth=2.5, markerfacecolor='white',
        markeredgecolor='green', label='marks',alpha=0.5)




average =sum(Marks)/len(Marks)
ax.axhline(y=average,
           color='navy',
           linestyle='--',
           linewidth=1.8,
           label=f'Average:{average:.1f}C')

ax.set_title("Student Marks ", fontsize=15, fontweight='bold')
ax.set_xlabel("Subjects", fontsize=12)
ax.set_ylabel("Marks", fontsize=12)
ax.legend(fontsize=11)

ax.grid(True, linestyle='--', alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()




```
