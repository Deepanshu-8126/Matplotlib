````python
# import matplotlib
import matplotlib.pyplot as plt
plt.plot([1,2,3,4]),[10,20,25,30]
plt.show()
````

````python
fig,ax=plt.subplots(figsize=(8,5))
# figsize=(width, height) in inches

# plot from axes
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_title(" My first Graph")
ax.set_xlabel("X - Axis Label")
ax.set_ylabel("Y - Axis Label")
plt.show()
````

````python
# professional touch
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 5))

# Data
x = [2019, 2020, 2021, 2022, 2023]
y = [100, 150, 130, 200, 250]
ax.plot(x, y, color='blue', linewidth=2, marker='o')
ax.set_title("Company Sales (2019-2023)", fontsize=14, fontweight='bold')
ax.set_xlabel("Year",fontsize=12,fontweight='bold')
ax.set_ylabel("Sales (in thousands)",fontsize=12)


# Grid
ax.grid(True , linestyle='--', alpha=0.6)
# alpha  transparency (0-1)


# automatically spacing fix
plt.tight_layout()
plt.show()
````

````python
# SINGLE LINE PLOt
import matplotlib.pyplot as plt
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']
temp = [15, 17, 22, 28, 35, 40, 38, 37, 32, 26, 19, 14]
fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(months,temp,color='tomato',linewidth=2.5,marker='o',markersize=8,markerfacecolor='white',  markeredgecolor='tomato',markeredgewidth=2,label='Tempreature')
ax.legend()


ax.set_title("Delhi Monthly Temperature", fontsize=15, fontweight='bold')
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Temperature (°C)", fontsize=12)

ax.grid(True, linestyle='--', alpha=0.5)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
