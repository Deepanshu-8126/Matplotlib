```python
# step 1 0 2 graphs side by side
import matplotlib.pyplot as plt

# Data
months = ['Jan','Feb','Mar','Apr','May','Jun']
sales  = [100, 120, 115, 140, 160, 150]
profit = [20,  35,  30,  50,  65,  55]

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,5))
#   ax1 - left grpah ax2 - right  graph

# left graph
ax1.plot(months, sales, color='royalblue',
         marker='o', linewidth=2)
ax1.set_title("Monthly Sales", fontsize=13, fontweight='bold')
ax1.set_xlabel("Month")
ax1.set_ylabel("Sales")
ax1.grid(True, linestyle='--', alpha=0.5)

# right graph

ax2.plot(months, profit, color='tomato',
         marker='s', linewidth=2)
ax2.set_title("Monthly Profit", fontsize=13, fontweight='bold')
ax2.set_xlabel("Month")
ax2.set_ylabel("Profit")
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# step 2 - 2 grpahs upside beside
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun']
sales  = [100, 120, 115, 140, 160, 150]
profit = [20,  35,  30,  50,  65,  55]
fig,(ax1,ax2)=plt.subplots(2,1,figsize=(9,8))

# 2 rows 1 column
ax1.bar(months, sales, color='royalblue',
        edgecolor='black', linewidth=0.7)
ax1.set_title("Monthly Sales", fontsize=13, fontweight='bold')
ax1.set_ylabel("Sales")
ax1.grid(True, axis='y', linestyle='--', alpha=0.5)


# beside graph
ax2.bar(months, profit, color='seagreen',
        edgecolor='black', linewidth=0.7)
ax2.set_title("Monthly Profit", fontsize=13, fontweight='bold')
ax2.set_xlabel("Month")
ax2.set_ylabel("Profit")
ax2.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

```

```python
# step 3 - 4 graphs at one plavce 2* 2grid

import matplotlib.pyplot as plt
import numpy as np

# Data
months    = ['Jan','Feb','Mar','Apr','May','Jun']
sales     = [100, 120, 115, 140, 160, 150]
profit    = [20, 35, 30, 50, 65, 55]
customers = [200, 240, 220, 280, 310, 290]
np.random.seed(42)
scores    = np.random.randint(40, 100, 50)



fig,axes = plt.subplots(2,2,figsize=(12,9))
# 2x2 grid — 4 graphs!
# axes = 2D array:
# axes[0][0] = top left
# axes[0][1] = top right
# axes[1][0] = bottom left
# axes[1][1] = bottom right

axes[0][0].plot(months, sales,
                color='royalblue', marker='o', linewidth=2)
axes[0][0].set_title("Sales Trend", fontweight='bold')
axes[0][0].set_ylabel("Sales")
axes[0][0].grid(True, linestyle='--', alpha=0.5)



# ── Top Right — Bar Chart ──
axes[0][1].bar(months, profit,
               color='tomato', edgecolor='black', linewidth=0.7)
axes[0][1].set_title("Monthly Profit", fontweight='bold')
axes[0][1].set_ylabel("Profit")
axes[0][1].grid(True, axis='y', linestyle='--', alpha=0.5)

# ── Bottom Left — Scatter Plot ──
axes[1][0].scatter(sales, profit,
                   color='seagreen', s=100,
                   edgecolor='black', alpha=0.8)
axes[1][0].set_title("Sales vs Profit", fontweight='bold')
axes[1][0].set_xlabel("Sales")
axes[1][0].set_ylabel("Profit")
axes[1][0].grid(True, linestyle='--', alpha=0.5)

# ── Bottom Right — Histogram ──
axes[1][1].hist(scores, bins=10,
                color='mediumpurple', edgecolor='black',
                linewidth=0.7, alpha=0.85)
axes[1][1].set_title("Score Distribution", fontweight='bold')
axes[1][1].set_xlabel("Score")
axes[1][1].set_ylabel("Count")
axes[1][1].grid(True, axis='y', linestyle='--', alpha=0.5)
# whole figure title suptitle
fig.suptitle("Sales Dashboard 2024",fontsize=18, fontweight='bold', y=1.01)

plt.tight_layout()
plt.show()
```

```python
# step 5 0 sharex and sharey pro
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun']
sales  = [100, 120, 115, 140, 160, 150]
profit = [20,  35,  30,  50,  65,  55]

fig, (ax1, ax2) = plt.subplots(2,1,figsize=(9,8),sharex=True)

# sharex = True both  2 x axis will same
# x axis label sonly shownn in down side one graph


ax1.plot(months, sales, color='royalblue',
         marker='o', linewidth=2)
ax1.set_title("Sales", fontweight='bold')
ax1.set_ylabel("Sales")
ax1.grid(True, linestyle='--', alpha=0.5)


ax2.bar(months, profit, color='tomato',
        edgecolor='black')
ax2.set_title("Profit", fontweight='bold')
ax2.set_xlabel("Month")
ax2.set_ylabel("Profit")
ax2.grid(True, axis='y', linestyle='--', alpha=0.5)

fig.suptitle("Sales & Profit Analysts", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

# sharex use when both graph has his x axis same data clean and professional

# sharex=True for x axis
# sharey= True for y axis
SIZE RULE                                             │
│  1×2 grid → figsize=(12, 5)                          │
│  2×1 grid → figsize=(8, 10)                          │
│  2×2 grid → figsize=(12, 9)
