Styling + Themes

- what comes in styling
1- themes - whole graph look change
2- colors - custom colors , colormaps
3 - spines - graph borders
4- ticks -axis numbers
5- legend -customize

```python
# step 1- themes ( one line of code look change)
import matplotlib.pyplot as plt

# available themes
print(plt.style.available)
```

```python
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun']
sales  = [100, 120, 115, 140, 160, 150]

# Theme implement
#plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(months, sales, color='royalblue',
        marker='o', linewidth=2.5)

ax.set_title("Sales with Dark Grid Theme",
             fontsize=15, fontweight='bold')
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Sales", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Step 2 Compare diffrent different themes
import matplotlib.pyplot as plt
months = ['Jan','Feb','Mar','Apr','May','Jun']
sales  = [100, 120, 115, 140, 160, 150]

# 4 themes compare
themes = ['default', 'seaborn-v0_8-darkgrid',
          'ggplot', 'dark_background']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i , theme in enumerate(themes):
  row = i // 2
  col = i%2
  with plt.style.context(theme):
   axes[row][col].plot(months, sales,
                            marker='o', linewidth=2)
   axes[row][col].set_title(f"Theme: {theme}",
                                  fontweight='bold')
  axes[row][col].grid(True)

plt.tight_layout()
plt.show()s
```

```python
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun']
sales  = [100, 120, 115, 140, 160, 150]

fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(months, sales, color='royalblue',
        marker='o', linewidth=2.5)

# ── Spines customize karo ──
# Graph ke 4 borders hote hain:
# top, bottom, left, right

ax.spines['top'].set_visible(False)
# ↑ upar wala border hatao

ax.spines['right'].set_visible(False)
# ↑ right wala border hatao

ax.spines['left'].set_color('grey')
# ↑ left border grey karo

ax.spines['bottom'].set_color('grey')
# ↑ bottom border grey karo

# Tick marks customize karo
ax.tick_params(axis='both',
# ↑ dono axis ke ticks
               labelsize=11,
               # ↑ tick numbers ka size
               colors='grey'
               # ↑ tick numbers ka color
               )

ax.set_title("Clean Minimal Style",
             fontsize=15, fontweight='bold')
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Sales", fontsize=12)
ax.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-darkgrid')

months   = ['Jan','Feb','Mar','Apr','May','Jun']
sales    = [100, 120, 115, 140, 160, 150]
profit   = [20, 35, 30, 50, 65, 55]

np.random.seed(42)
scores   = np.random.randint(40, 100, 50)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# ── Graph 1 — Line Plot ──
axes[0].plot(months, sales,
             color='#2ECC71', marker='o',
             linewidth=2.5, markersize=8,
             markerfacecolor='white',
             markeredgecolor='#2ECC71',
             markeredgewidth=2)
axes[0].set_title("Sales Trend", fontsize=13,
                   fontweight='bold')
axes[0].set_ylabel("Sales (units)")
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

# ── Graph 2 — Bar Chart ──
bar_colors = ['#E74C3C' if p < 40 else '#3498DB'
              for p in profit]
# ↑ 40 se kam = red, zyada = blue
axes[1].bar(months, profit,
            color=bar_colors,
            edgecolor='black', linewidth=0.6)
axes[1].set_title("Monthly Profit", fontsize=13,
                   fontweight='bold')
axes[1].set_ylabel("Profit (units)")
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

# ── Graph 3 — Histogram ──
axes[2].hist(scores, bins=12,
             color='#9B59B6',
             edgecolor='black',
             linewidth=0.6, alpha=0.85)
avg = np.mean(scores)
axes[2].axvline(x=avg, color='red',
                linestyle='--', linewidth=2,
                label=f'Avg: {avg:.1f}')
axes[2].set_title("Score Distribution", fontsize=13,
                   fontweight='bold')
axes[2].set_ylabel("Count")
axes[2].legend()
axes[2].spines['top'].set_visible(False)
axes[2].spines['right'].set_visible(False)

fig.suptitle("Business Analytics Dashboard",
             fontsize=18, fontweight='bold', y=1.02)

plt.tight_layout()
plt.show()
```
