```python
# step 2  percentage shown
import matplotlib.pyplot as plt

brands = ['Android', 'iPhone', 'Others']
market = [72, 18, 10]

fig, ax = plt.subplots(figsize=(7, 7))
# autopct - automatic percentage show %1 - pecertage sign 1.1f - 1 decimal place % sign show
# %% - percentage sign shown


ax.pie(market, labels=brands,autopct='%1.1f%%')
ax.set_title("Phone Market Share", fontsize=15, fontweight='bold')
plt.show()

"""#**Dry Run — autopct:**

market = [72, 18, 10]
total  = 100

Android → 72/100 × 100 = 72.0%
iPhone  → 18/100 × 100 = 18.0%
Others  → 10/100 × 100 = 10.0%

Graph pe dikhega:
Android 72.0%
iPhone  18.0%
Others  10.0%  ✅"""

```

```python
# colour + styling
import matplotlib.pyplot as plt

brands = ['Android', 'iPhone', 'Others']
market = [72, 18, 10]

# colours define
colors = ['blue', 'green','tomato']


# startangle - where to start circle 90 clock starting 12
fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(market,
       labels=brands,
       colors=colors,
       autopct='%1.1f%%',
       startangle = 90,
       wedgeprops = {'edgecolor':'white', 'linewidth':2}
       # wedgeprops - all slice properties
       # edgecolor = between slices white line
       )

ax.set_title("Phone market share 2024 ", fontsize = 15, fontweight = 'bold')

plt.show()
```

```python
# step 4 - explode ( one slice take outside )
import matplotlib.pyplot as plt

brands = ['Android', 'iPhone', 'Others']
market = [72, 18, 10]
colors = ['#3DDC84', '#555555', '#A8A8A8']

# explode - how many want outside
explode = (0,0.1,0)
#        android , iphone , others
#  0 - inside -- 0.1 - outside  next zero 0 - inside
# only i phone slice comes outside

fig, ax = plt.subplots(figsize=(7, 7))

ax.pie(market,
       labels=brands,
       colors=colors,
       autopct='%1.1f%%',
       startangle=90,
       explode=explode,
       shadow = True,
       wedgeprops={'edgecolor': 'white', 'linewidth': 2}
       )
ax.set_title("Phone Market Share — iPhone Highlighted",
             fontsize=15, fontweight='bold')
plt.show()
```

```python
# step 5 - donut chart
import matplotlib.pyplot as plt

# Budget breakdown
categories = ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings']
budget     = [40, 25, 10, 15, 10]
colors     = ['#E74C3C', '#3498DB', '#2ECC71', '#F39C12', '#9B59B6']
fig, ax = plt.subplots(figsize=(8, 8))


# wedges - slice objects
# texts - label text objects
# autotexts - percentage text objects
wedges , texts , autotexts = ax.pie(
     budget,
    labels=categories,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={
        'edgecolor': 'white',
        'linewidth': 2,
        'width': 0.6

    }

)

# percentage text style
for autotext in autotexts:
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')


# text at the center (0,0) - center
ax.text(0,0,'Budget\n2026',
        ha='center',
        va='center',
        fontsize=16,

      fontweight='bold')
ax.set_title("Monthly Budget Breakdown", fontsize=15, fontweight='bold')
plt.show()

```
